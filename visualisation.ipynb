if the population of Oman in 2023 is 4644384 and the growth rate is 1.49%
What will be the population in 2040

```python
p = 4644384
g = 1.49/100
for y in range(2023,2041):
    p = p+p*g
    print (y, p)
print ("The population in 2040 is: ", p)
```

if the balance of a bank customer is 1000, and the interest rate is 3.5% find the following:
1- the balance after 4 years
2- how many years does it take to reach 10,000

```python
a = 1000
r = 3.5/100
for i in range(0,4):
    a = a +a*r
    print (i, a)
print ("The amount after 4 years is: ", a)
```
