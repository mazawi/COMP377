draw the relationship between the expenses and the month

```python
Month = ["Jan", 'Feb', 'March', 'April']
Expe = [200, 300, 500, 400]
income = [400, 500, 600, 500]
plt.plot(Month, Expe, color = 'y')
plt.title("Expenses per month")
plt.xlabel("Monthes")
plt.ylabel("Expenses (OMR)")

plt.plot(Month, income)
```

```python
plt.bar(Month, Expe)
```

```python
plt.barh(Month, Expe)
```

```python
plt.pie(Expe, labels = Month)
```
